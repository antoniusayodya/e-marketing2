The most considered factors in selecting any dating social media to subscribe.

```python
variabel = {'keanggotaan':['24','26'],
           'kualitas_layanan':['27', '31'],
           'harga_berlangganan': ['32', '34'],
           'kualitas_laman_atau_app': ['35', '41'],
           'proteksi_privasi_data' : ['42', '42'],
           'layanan pelanggan' : ['43', '44'],
           'promosi' : ['45', '45']}

for kriteria, batasan in variabel.items():
    df[kriteria] = df.loc[:, batasan[0]:batasan[1]].mean(axis = 1)
df3 = df.loc[:, 'keanggotaan':'promosi'].mean()
df3 = df3.to_frame(name = 'rerata')
df3.sort_values('rerata', inplace = True, ascending = False)
df3.plot.bar()
```
